```python
import os
import numpy as np 
import pandas as pd
from PIL import Image

import plotly 
import plotly.express as px
import plotly.graph_objs as go
```

```python
colordb = pd.read_csv("../backend/colordb.csv")
```

```python
img_path = "../public/miyuki.png"
```

```python
def extract_flat_color_array(img_path):
    def _clip_alpha(img_array):
        is_alpha = img_array.shape[-1] == 4
        return img_array[:, :, :3] if is_alpha else img_array

    img = Image.open(img_path)
    m, n = img.height, img.width
    img_resized = img.resize((m//10, n//10))
    img_array = np.array(img_resized)
    img_array = _clip_alpha(img_array)
    print(img_array.shape)
    flat_colors_array = img_array.reshape((img_array.shape[0]*img_array.shape[1], 3))
    
    return flat_colors_array
    
def filter_unique_colors(flat_colors_array):
    unique_colors = list(set(tuple(vec) for vec in flat_colors_array))
    unique_colors = np.array(unique_colors)        
    
    return unique_colors 
```

```python
flat_colors_array = extract_flat_color_array(img_path)
```

```python
unique_colors = filter_unique_colors(flat_colors_array)
```

```python
r = unique_colors[:, 0]
g = unique_colors[:, 1]
b = unique_colors[:, 2]

trace = go.Scatter3d(
            x=r, y=g, z=b,
            mode='markers',
            marker=dict(
                size=2,
                color=[
                    'rgb({},{},{})'.format(r,g,b) 
                    for r,g,b in zip(r, g, b)
                    ],
                opacity=0.9),
            hovertext=colordb['color_name']
        )
```

```python
scatter = [trace]
```

```python
layout = go.Layout(margin=dict(l=0, r=0, b=0, t=0))
```

```python
fig = go.Figure(data=scatter)
fig.update_layout(
    plot_bgcolor='rgb(17,17,17)',
    paper_bgcolor ='rgb(10,10,10)')
```

```python

fig.show()
```
